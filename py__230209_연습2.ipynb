```python
import requests  # requests 라이브러리 불러오기
res = requests.get ('http://ggoreb.com/api/shop.jsp') #사이트에 있는 데이터 불러오기
result = res.json() #리스트, 딕셔너리 변환
```

```python
list = result.get('list') #'list'대가로 안에 있는 데이터
for item in list : #'list'안에 lng, lat 데이터 for문이용해서 출력
    lat = item.get('lat')
    lng = item.get('lng')
    print(lat,lng)
```
