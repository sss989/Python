# 로또 일등 번호 가져오기

```python
import requests
res = requests.get('https://dhlottery.co.kr/gameResult.do?method=byWin')
result = res.text
```

```python
result
```

```python
sidx = result.find('<strong>1053회</strong>')
eidx = result.find('num bonus', sidx)
result2 = result[sidx : eidx]
print(result2)
```

```python
# 번호 추출 / 슬라이싱 해서 추출

# sidx = result2.find('ball_645') # 이거 먼저 치면 '22'밖에 안나옴

sidx = result2.find('ball_645', eidx) # 이렇게 하면 다음 값들이 나옴


sidx = result2.find('>', sidx)
eidx = result2.find('<', sidx)
result2[sidx + 1 : eidx]
```

```python
eidx = 0
while True :
    sidx = result2.find('ball_645', eidx)
    if sidx == -1: break
    sidx = result2.find('>', sidx)
    eidx = result2.find('<', sidx)
    print(result2[sidx + 1 : eidx])
```

##### 점을 안찍는다 == 기본 자료형

```python
num = 123 # 1 2 3
일자리 = num % 10
print(일자리)

십자리 = num // 10% 10
print(십자리)

백자리 = num // 100% 10
print(백자리)
```
