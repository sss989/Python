```python
import requests
res = requests.get('http://ggoreb.com/dinosaur/info.jsp')
```

```python
result = res.json() # 많은 데이터를 딕셔너리, 리스트 형태로 변환해줌.
```

```python
res.text # 문자형태 그대로 나옴
```

```python
result
```

```python
length = len(result[0]['dinosaurs'][0]['imgs'])
for i in range(0, length):
    print(i,result[0]['dinosaurs'][0]['imgs'][i])
```

```python
for value in result[0]['dinosaurs'][0]['imgs']:
    print(value)
```

```python
result[0]['dinosaurs'][0]['title']
```

```python
for re in result :
    kind = re['kind']
    print(kind)  # ,get 쓰면 없어도 에러가 나지않음.
```

```python
# result => 전체 리스트 [{},{}]
for dinosaur in result :
    dinosaurs = dinosaur.get('dinosaurs')
    for dino in dinosaurs :
        title = dino.get('title')
        print(title)
```
